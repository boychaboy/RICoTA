````python
def draw_confusion_matrix(gold, pred, class_names):
    # gold = conversation_type['gold']
    # pred = conversation_type['chatgpt']
    
    # class_names = conversation_types
    
    cm = confusion_matrix(gold, pred, labels=class_names)
    
    plt.figure(figsize=(15, 10))
    plt.xticks(rotation=45)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)    
````
